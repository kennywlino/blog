The next step we want to do in order to count the words is to actually *separate* them. What do I mean by that? Well, we need a way for our tools to distinguish one word from another. The easiest way here is to put each word on its own line.

To do this in bash, we'll use the translate tool, or `tr` to replace all the whitespace with a newline character. Because there are a lot of possible whitespace characters, we'll take the opposite or, the *complement* of anything that's not A-Z.

```python
#collapse-output
! tr -sc '[A-Z][a-z]' '\n' < ../../01_the_ugly_duckling.txt
```
